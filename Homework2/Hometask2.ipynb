### __*Пункт 1:*__

Составим текст из разных предложений с разными случаями омоформ (например: бурят (глаг) и бурят (сущ)). Есть случаи последовательного употребления омоформ ("На оконное стекло стекло несколько капель дождя"). Данные случаи достаточно трудны для распознавания. В качестве основного тегсета возмем тот, который используется в pymorphy.

```python
with open('HW2 - text.txt', encoding='utf-8') as file:
    text = file.read()
```

### __*Пункт 2:*__

Первым pos-теггером будет pymorphy. Вытащим для кажого слова из текста его часть речи и добавим в список. 

```python
from pymorphy2 import MorphAnalyzer
m = MorphAnalyzer()
```

```python
pm_taggs = []
from pymorphy2.tokenizers import simple_word_tokenize
for token in simple_word_tokenize(text):
    h = m.parse(token)[0]
    if h.tag.POS != None:
        pm_taggs.append([h.word, h.tag.POS])
```

```python
pm_taggs
```

Вторым будет mystem. Также сразу подгоним наши теги под используемый стандарт. 

```python
from pymystem3 import Mystem
mystem = Mystem()

a_text = mystem.analyze(text)
```

```python
ms_taggs = []
for word in a_text:
    if 'analysis' in word:
        gr = word['analysis'][0]['gr']
        pos = gr.split('=')[0].split(',')[0]
        if pos == 'S':
            ms_taggs.append([word['text'], "NOUN"])
        elif pos == 'V':
            ms_taggs.append([word['text'], "VERB"])
        elif pos == 'PR':
            ms_taggs.append([word['text'], "PREP"])
        elif pos == 'PART':
            ms_taggs.append([word['text'], "PRCL"])
        elif pos == 'A' or pos == 'APRO':
            ms_taggs.append([word['text'], "ADJF"])
        elif pos == 'ADV' or pos == 'ADVPRO':
            ms_taggs.append([word['text'], "ADVB"])
        elif pos == 'SPRO':
            ms_taggs.append([word['text'], "NPRO"])
        else:
            ms_taggs.append([word['text'], pos])
            
```

```python
ms_taggs
```

Третим будет stanza. Скачиваем документацию для русского языка, аналогично стандартизируем теги. 

```python
import stanza

stanza.download('ru')
nlp = stanza.Pipeline('ru')
```

```python
doc = nlp(text)
stan_taggs = []

for sentence in doc.sentences:
    for word in sentence.words:
        if word.pos != 'PUNCT':
            if word.pos == 'PROPN':
                stan_taggs.append([word.text, 'NOUN'])
            elif word.pos == 'ADP':
                stan_taggs.append([word.text, 'PREP'])
            elif word.pos == 'AUX':
                stan_taggs.append([word.text, 'VERB'])
            elif word.pos == 'SCONJ' or word.pos == 'CCONJ':
                stan_taggs.append([word.text, 'CONJ'])
            elif word.pos == 'PART':
                stan_taggs.append([word.text, 'PRCL'])
            elif word.text != '―':
                stan_taggs.append([word.text, word.pos])
```

```python
stan_taggs
```

### __*Пункт 4:*__

Напишем три функции, чтобы выделаить три разных набора n-грамм. Первая: найдем все NP.

```python
def chunker1():
    ngramm_NP = []
    for i in range(0, 207):
        if ms_taggs[i][1] == 'NOUN':
            if ms_taggs[i-1][1] == 'ADJF':
                ngramm_NP.append([ms_taggs[i-1][0], ms_taggs[i][0]])
    print(ngramm_NP)
```

```python
chunker1()
```

Вторая: найдем все выражения с отрицательной частицей.

```python
def chunker2():
    ngramm_no = []
    for j in range(0, 207):
        if ms_taggs[j][1] == 'PRCL':
            if ms_taggs[j+1][1] == 'VERB':
                ngramm_no.append([ms_taggs[j][0], ms_taggs[j+1][0]])
    print(ngramm_no)
```

```python
chunker2()
```

Третья: найдем все последовательности сущ - глаг - сущ.

```python
def chunker3():
    ngramm_NVN = []
    for n in range(0, 206):
        if ms_taggs[n][1] == 'NOUN':
            if ms_taggs[n+1][1] == 'VERB':
                if ms_taggs[n+2][1] == 'NOUN':
                    ngramm_NVN.append([ms_taggs[n][0], ms_taggs[n+1][0], ms_taggs[n+2][0]])
    print(ngramm_NVN)
```
